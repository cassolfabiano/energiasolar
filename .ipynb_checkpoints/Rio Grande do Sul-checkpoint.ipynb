# Processamento de mapas

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
CIDADES = [
    "SANTO AUGUSTO",
    "SANTA ROSA",
    "SÃO LUIZ GONZAGA",
    "CRUZ ALTA",
]
```

```python
SantaRosa = pd.read_excel("RS_A810_SANTA_ROSA_t_ur_rad_prec_2014_18.xlsx")
SantoAugusto = pd.read_excel("RS_A805_SANTO_AUGUSTO_t_ur_rad_prec_2014_18.xlsx")
CruzAlta = pd.read_excel("RS_A853_CRUZ_ALTA_t_ur_rad_prec_2014_18.xlsx")
SaoLuizGonzaga = pd.read_excel("RS_A852_SAO_LUIZ_GONZAGA_t_ur_rad_prec_2014_18.xlsx")
```

```python
RS = gpd.read_file("rs_municipios/43MUE250GC_SIR.shp", encoding="utf-8")
RS_filtered = gpd.GeoDataFrame(RS[RS.NM_MUNICIP.isin(CIDADES)])
RS_filtered.loc[370, ("Tar")] = SantaRosa[("tar")].min()
RS_filtered.loc[380, ("Tar")] = SantoAugusto[("tar")].min()
RS_filtered.loc[124, ("Tar")] = CruzAlta[("tar")].min()
RS_filtered.loc[402, ("Tar")] = SaoLuizGonzaga[("tar")].min()
RS_filtered
```

```python
fig, ax = plt.subplots(figsize=(16,16))
RS_filtered.plot(column="Tar", cmap="viridis", edgecolor='k', ax=ax)
```
